````python

"""
Agente Conversacional usando LangGraph y Ollama
==============================================

Este c√≥digo implementa un agente conversacional capaz de tomar decisiones sobre c√≥mo responder
a las consultas del usuario, incluyendo el uso de herramientas cuando es necesario.

Al final de cada interacci√≥n, genera un diagrama de flujo que muestra la ruta exacta tomada
desde la consulta del usuario hasta la respuesta final.
"""

#==================================================
# 1. CONFIGURACI√ìN INICIAL
#==================================================

import os
import re
import time
import logging
from typing import TypedDict, Annotated, List, Literal, Union, Dict, Any, Optional

# Importaciones de LangChain y LangGraph
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage, FunctionMessage
from langchain_core.tools import tool
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langgraph.graph import StateGraph, END

# Importaciones para visualizaci√≥n
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib

# Configurar el sistema de logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.StreamHandler()  # Salida a consola
    ]
)
logger = logging.getLogger("AgenteLangGraph")


#==================================================
# 2. DEFINICI√ìN DE TIPOS Y ESTADO
#==================================================

# Definir el estado del agente conversacional
class ConversationalAgentState(TypedDict):
    messages: List[Annotated[Union[SystemMessage, HumanMessage, AIMessage, FunctionMessage], "Historial de la conversaci√≥n"]]
    next_step: Literal["decide_action", "generate_response", "use_calculator", "end"]

# Registro de la ruta tomada durante la ejecuci√≥n
execution_path = []

# Funci√≥n para formatear el estado para logs
def format_state_for_log(state):
    """Formatea el estado del agente para mostrar en logs"""
    messages = []
    for msg in state.get("messages", []):
        msg_type = type(msg).__name__
        if isinstance(msg, SystemMessage):
            content = msg.content[:50] + "..." if len(msg.content) > 50 else msg.content
            messages.append(f"[{msg_type}]: {content}")
        else:
            content = msg.content[:50] + "..." if len(msg.content) > 50 else msg.content
            messages.append(f"[{msg_type}]: {content}")
    
    formatted = {
        "message_count": len(messages),
        "message_types": messages,
        "next_step": state.get("next_step", None)
    }
    
    # A√±adir otros campos si existen
    for key in state:
        if key not in ["messages", "next_step"]:
            formatted[key] = state[key]
            
    return formatted


#==================================================
# 3. CONFIGURACI√ìN DEL MODELO LLM
#==================================================

# Configurar Ollama con el modelo deepseek-r1:1.5b:32b
llm = Ollama(model="deepseek-r1:1.5b:32b")
logger.info(f"Modelo Ollama inicializado: deepseek-r1:1.5b:32b")

# Sistema de instrucciones para el agente con informaci√≥n sobre herramientas
SYSTEM_PROMPT = """Eres un asistente conversacional amigable y servicial con acceso a herramientas.

Tienes acceso a las siguientes herramientas:

1. calculadora: Puedes usar esta herramienta para realizar c√°lculos matem√°ticos precisos.
   Ejemplos de uso:
   - Para sumar n√∫meros: calculadora("2 + 2")
   - Para multiplicar: calculadora("5 * 3")
   - Para dividir: calculadora("10 / 2")
   - Para operaciones m√°s complejas: calculadora("(5 + 3) * 2")

Para usar una herramienta, analiza cuidadosamente si la pregunta del usuario requiere c√°lculos matem√°ticos.
Si es as√≠, DEBES indicar que usar√°s la calculadora y proporcionar la expresi√≥n adecuada.

Si la consulta no requiere c√°lculos, simplemente responde directamente sin usar herramientas.

S√© amable, detallado y siempre intenta ser de ayuda. Si no sabes algo, simplemente adm√≠telo en lugar de inventar informaci√≥n."""

logger.info("Sistema de prompt configurado")


#==================================================
# 4. HERRAMIENTAS DEL AGENTE
#==================================================

# Definici√≥n de la herramienta de calculadora
@tool
def calculadora(operacion: str) -> str:
    """
    Realiza c√°lculos matem√°ticos.
    
    Args:
        operacion: Expresi√≥n matem√°tica como "2 + 2", "5 * 3", "10 / 2", etc.
        
    Returns:
        El resultado del c√°lculo.
    """
    logger.info(f"Calculadora recibi√≥ operaci√≥n: {operacion}")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "calculadora",
        "description": f"Calculando: {operacion}"
    })
    
    try:
        # Limpiamos la expresi√≥n para mayor seguridad
        operacion = operacion.strip()
        # Evaluamos solo si contiene operaciones matem√°ticas v√°lidas
        if re.match(r'^[\d\s\+\-\*\/\(\)\.\,\^\%]+$', operacion):
            # Reemplazamos ^ por ** para potencias
            operacion = operacion.replace('^', '**')
            resultado = eval(operacion)
            logger.info(f"Calculadora complet√≥: {operacion} = {resultado}")
            
            # Registrar resultado
            execution_path.append({
                "node": "calculadora_resultado",
                "description": f"Resultado: {resultado}"
            })
            
            return f"El resultado de {operacion} es {resultado}"
        else:
            logger.warning(f"Calculadora rechaz√≥ operaci√≥n con caracteres no permitidos: {operacion}")
            
            # Registrar error
            execution_path.append({
                "node": "calculadora_error",
                "description": "Error: Caracteres no permitidos"
            })
            
            return "La operaci√≥n contiene caracteres no permitidos. Solo puedo realizar operaciones con n√∫meros y los operadores +, -, *, /, (, ), ., ^, %"
    except Exception as e:
        logger.error(f"Error en calculadora: {str(e)}, operaci√≥n: {operacion}")
        
        # Registrar error
        execution_path.append({
            "node": "calculadora_error",
            "description": f"Error: {str(e)}"
        })
        
        return f"Error al calcular: {str(e)}"


#==================================================
# 5. NODOS DEL GRAFO
#==================================================

# Funci√≥n para decidir la acci√≥n a tomar
def decide_action(state: ConversationalAgentState) -> ConversationalAgentState:
    """Decide si responder directamente o usar una herramienta"""
    logger.info(f"Iniciando decide_action con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "decide_action",
        "description": "Analizando input del usuario"
    })
    
    # Obtener el √∫ltimo mensaje del usuario
    last_message = state["messages"][-1]
    
    if not isinstance(last_message, HumanMessage):
        # Si no es un mensaje humano, pasamos a generar respuesta
        logger.info("El √∫ltimo mensaje no es humano, pasando a generate_response")
        
        execution_path.append({
            "node": "decide_action_result",
            "description": "No es mensaje humano, pasando a generate_response"
        })
        
        state["next_step"] = "generate_response"
        return state
    
    # Crear un prompt para decidir la acci√≥n
    history = state["messages"]
    context = "\n".join([
        f"{'Usuario' if isinstance(msg, HumanMessage) else 'Asistente' if isinstance(msg, AIMessage) else 'Sistema' if isinstance(msg, SystemMessage) else 'Funci√≥n'}: {msg.content}" 
        for msg in history[-5:] if not isinstance(msg, SystemMessage) or msg == history[0]
    ])
    
    decision_prompt = f"""
{SYSTEM_PROMPT}

Historial reciente:
{context}

Decide qu√© acci√≥n tomar con la √∫ltima consulta del usuario:
1. Si requiere un c√°lculo matem√°tico, escribe exactamente 'USAR_CALCULADORA: <expresi√≥n>' donde <expresi√≥n> es la operaci√≥n matem√°tica.
2. Para cualquier otra consulta, escribe exactamente 'RESPONDER_DIRECTAMENTE'.

Tu decisi√≥n (solo escribe USAR_CALCULADORA o RESPONDER_DIRECTAMENTE):
"""
    
    # Obtener la decisi√≥n del modelo
    logger.info(f"Solicitando decisi√≥n al modelo para: {last_message.content}")
    
    execution_path.append({
        "node": "llm_deciding",
        "description": f"LLM analizando: '{last_message.content}'"
    })
    
    decision = llm.invoke(decision_prompt).strip()
    logger.info(f"Decisi√≥n del modelo: {decision}")
    
    # Analizar la decisi√≥n
    if "USAR_CALCULADORA:" in decision:
        # Extraer la expresi√≥n matem√°tica
        match = re.search(r'USAR_CALCULADORA:\s*(.*)', decision)
        if match:
            expression = match.group(1).strip()
            logger.info(f"Se usar√° la calculadora con expresi√≥n: {expression}")
            
            # Registrar decisi√≥n
            execution_path.append({
                "node": "decide_action_result",
                "description": f"Usar calculadora con expresi√≥n: {expression}"
            })
            
            # Guardar la expresi√≥n en el estado para usarla despu√©s
            state["expression"] = expression
            state["next_step"] = "use_calculator"
        else:
            # Si no podemos extraer la expresi√≥n, responder directamente
            logger.warning("No se pudo extraer la expresi√≥n para la calculadora")
            
            # Registrar decisi√≥n
            execution_path.append({
                "node": "decide_action_result",
                "description": "Error al extraer expresi√≥n, respondiendo directamente"
            })
            
            state["next_step"] = "generate_response"
    else:
        # Responder directamente si no es un c√°lculo
        logger.info("Se responder√° directamente sin usar herramientas")
        
        # Registrar decisi√≥n
        execution_path.append({
            "node": "decide_action_result",
            "description": "Responder directamente sin herramientas"
        })
        
        state["next_step"] = "generate_response"
    
    logger.info(f"Estado despu√©s de decide_action: {format_state_for_log(state)}")
    return state

# Funci√≥n para usar la calculadora
def use_calculator(state: ConversationalAgentState) -> ConversationalAgentState:
    """Usa la herramienta de calculadora"""
    logger.info(f"Iniciando use_calculator con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "use_calculator",
        "description": "Ejecutando herramienta calculadora"
    })
    
    # Obtener la expresi√≥n a calcular
    expression = state.get("expression", "")
    
    # Usar la herramienta
    logger.info(f"Ejecutando calculadora con expresi√≥n: {expression}")
    result = calculadora(expression)
    logger.info(f"Resultado de la calculadora: {result}")
    
    # Agregar el resultado como un mensaje de funci√≥n
    state["messages"].append(FunctionMessage(content=result, name="calculadora"))
    
    # Registrar finalizaci√≥n
    execution_path.append({
        "node": "use_calculator_complete",
        "description": f"Completado: {result}"
    })
    
    # Pasar a generar la respuesta
    state["next_step"] = "generate_response"
    
    logger.info(f"Estado despu√©s de use_calculator: {format_state_for_log(state)}")
    return state

# Funci√≥n para generar respuesta usando Ollama
def generate_response(state: ConversationalAgentState) -> ConversationalAgentState:
    """Genera una respuesta conversacional usando el modelo de Ollama"""
    logger.info(f"Iniciando generate_response con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "generate_response",
        "description": "Generando respuesta al usuario"
    })
    
    # Construir el prompt con todo el historial de mensajes
    history = state["messages"]
    
    # Extraer el contenido de los mensajes para crear un contexto
    context = "\n".join([
        f"{'Usuario' if isinstance(msg, HumanMessage) else 'Asistente' if isinstance(msg, AIMessage) else 'Sistema' if isinstance(msg, SystemMessage) else f'Funci√≥n ({msg.name})'}: {msg.content}" 
        for msg in history if msg != history[0]  # Excluimos el primer mensaje del sistema
    ])
    
    # Crear el prompt para el modelo
    prompt = f"{SYSTEM_PROMPT}\n\nHistorial de conversaci√≥n:\n{context}\n\nAsistente:"
    
    # Obtener la respuesta del modelo
    logger.info("Solicitando respuesta al modelo")
    
    execution_path.append({
        "node": "llm_generating",
        "description": "LLM generando respuesta final"
    })
    
    response = llm.invoke(prompt)
    logger.info(f"Respuesta generada: {response[:100]}...")
    
    # Agregar la respuesta al estado
    state["messages"].append(AIMessage(content=response))
    state["next_step"] = "end"  # Marcar que hemos terminado este turno
    
    # Registrar finalizaci√≥n
    execution_path.append({
        "node": "generate_response_complete",
        "description": f"Respuesta generada: {response[:50]}..." if len(response) > 50 else response
    })
    
    logger.info(f"Estado despu√©s de generate_response: {format_state_for_log(state)}")
    return state

# Funci√≥n para decidir el siguiente paso
def router(state: ConversationalAgentState) -> Literal["decide_action", "generate_response", "use_calculator", "end"]:
    next_step = state["next_step"]
    logger.info(f"Router decidiendo siguiente paso: {next_step}")
    
    # Registrar decisi√≥n del router
    execution_path.append({
        "node": "router",
        "description": f"Siguiente paso: {next_step}"
    })
    
    return next_step


#==================================================
# 6. CONSTRUCCI√ìN DEL GRAFO
#==================================================

# Crear el grafo del agente conversacional
def build_conversational_agent():
    logger.info("Construyendo el grafo del agente conversacional")
    
    # Inicializar el grafo con el tipo de estado
    workflow = StateGraph(ConversationalAgentState)
    
    # Agregar nodos
    workflow.add_node("decide_action", decide_action)
    workflow.add_node("generate_response", generate_response)
    workflow.add_node("use_calculator", use_calculator)
    logger.info("Nodos a√±adidos al grafo: decide_action, generate_response, use_calculator")
    
    # Establecer el nodo de inicio
    workflow.set_entry_point("decide_action")
    logger.info("Punto de entrada establecido: decide_action")
    
    # Conectar los nodos seg√∫n la decisi√≥n
    workflow.add_conditional_edges("decide_action", router, {
        "generate_response": "generate_response",
        "use_calculator": "use_calculator"
    })
    logger.info("Bordes condicionales a√±adidos desde decide_action")
    
    # Despu√©s de usar la calculadora, generar respuesta
    workflow.add_conditional_edges("use_calculator", router, {
        "generate_response": "generate_response"
    })
    logger.info("Bordes condicionales a√±adidos desde use_calculator")
    
    # Despu√©s de generar respuesta, terminar
    workflow.add_conditional_edges("generate_response", router, {
        "end": END
    })
    logger.info("Bordes condicionales a√±adidos desde generate_response")
    
    # Compilar el grafo
    logger.info("Compilando el grafo")
    return workflow.compile()


#==================================================
# 7. VISUALIZACI√ìN DEL FLUJO DE TRABAJO
#==================================================

def generar_diagrama_flujo(query, response):
    """
    Genera un diagrama del flujo del agente basado en la ruta de ejecuci√≥n.
    
    Args:
        query: La consulta del usuario
        response: La respuesta final generada
    """
    logger.info("Generando diagrama del flujo de trabajo")
    
    # Crear un nuevo grafo para la visualizaci√≥n
    G = nx.DiGraph()
    
    # A√±adir nodos especiales para consulta y respuesta
    G.add_node("user_query", pos=(-2, 0), node_type="user", label=f"Consulta:\n{query[:30]}...")
    G.add_node("agent_response", pos=(4, 0), node_type="response", label=f"Respuesta:\n{response[:30]}...")
    
    # A√±adir nodos principales del agente
    G.add_node("decide_action", pos=(0, 0), node_type="decision", label="Decisi√≥n")
    G.add_node("use_calculator", pos=(2, 1), node_type="tool", label="Calculadora")
    G.add_node("generate_response", pos=(2, -1), node_type="generation", label="Generaci√≥n")
    
    # A√±adir aristas b√°sicas
    G.add_edge("user_query", "decide_action", label="Input")
    G.add_edge("decide_action", "use_calculator", label="Requiere\nherramienta")
    G.add_edge("decide_action", "generate_response", label="Respuesta\ndirecta")
    G.add_edge("use_calculator", "generate_response", label="Con resultado")
    G.add_edge("generate_response", "agent_response", label="Output")
    
    # Determinar la ruta tomada basada en execution_path
    path_edges = []
    route_type = "direct"  # Por defecto, asumimos ruta directa
    
    # Analizar el execution_path para determinar la ruta tomada
    for i in range(len(execution_path)):
        node = execution_path[i]["node"]
        
        if node == "decide_action_result":
            desc = execution_path[i]["description"].lower()
            if "usar calculadora" in desc or "use_calculator" in desc:
                path_edges.append(("decide_action", "use_calculator"))
                route_type = "with_tool"
            else:
                path_edges.append(("decide_action", "generate_response"))
                route_type = "direct"
    
    # Completar la ruta bas√°ndonos en el tipo de ruta detectado
    if route_type == "with_tool":
        path_edges.append(("use_calculator", "generate_response"))
    
    # A√±adir aristas de inicio y fin
    path_edges.insert(0, ("user_query", "decide_action"))
    path_edges.append(("generate_response", "agent_response"))
    
    # Crear figura
    plt.figure(figsize=(12, 8))
    
    # Posiciones de los nodos
    pos = nx.get_node_attributes(G, 'pos')
    
    # Colores para los nodos seg√∫n su tipo
    node_colors = {
        'user': '#66c2a5',        # Verde azulado para usuario
        'decision': '#fc8d62',    # Naranja para decisi√≥n
        'tool': '#8da0cb',        # Azul para herramienta
        'generation': '#e78ac3',  # Rosa para generaci√≥n
        'response': '#a6d854'     # Verde para respuesta
    }
    
    # Preparar colores para cada nodo
    color_map = [node_colors[G.nodes[node].get('node_type', 'decision')] for node in G.nodes()]
    
    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, node_size=2500, node_color=color_map, alpha=0.8, 
                          node_shape='o', edgecolors='black', linewidths=2)
    
    # Preparar etiquetas de nodos (con formato)
    node_labels = {}
    for node in G.nodes():
        label = G.nodes[node].get('label', node)
        node_labels[node] = label
    
    # Dibujar etiquetas de nodos
    nx.draw_networkx_labels(G, pos, node_labels, font_size=10, font_weight='bold')
    
    # Dibujar todas las aristas en gris claro (fondo)
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=1.5, 
                          edge_color='lightgray', style='dashed', 
                          arrows=True, arrowsize=15, 
                          connectionstyle='arc3,rad=0.1')
    
    # Dibujar la ruta tomada en color rojo y m√°s ancha
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, width=3.0, 
                          edge_color='red', style='solid', 
                          arrows=True, arrowsize=20, 
                          connectionstyle='arc3,rad=0.1')
    
    # Etiquetas de aristas
    edge_labels = {}
    for u, v, data in G.edges(data=True):
        if 'label' in data:
            edge_labels[(u, v)] = data['label']
    
    # Dibujar etiquetas de aristas
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, 
                               font_weight='bold', label_pos=0.5, rotate=False)
    
    # A√±adir t√≠tulo
    plt.title("Diagrama de Flujo del Agente", fontsize=16, fontweight='bold')
    
    # Quitar ejes
    plt.axis('off')
    
    # A√±adir una leyenda para explicar el flujo
    if route_type == "with_tool":
        route_description = "Ruta: Consulta ‚Üí Decisi√≥n ‚Üí Usar Calculadora ‚Üí Generar Respuesta ‚Üí Respuesta Final"
    else:
        route_description = "Ruta: Consulta ‚Üí Decisi√≥n ‚Üí Generar Respuesta ‚Üí Respuesta Final"
    
    plt.figtext(0.5, 0.01, route_description, ha='center', fontsize=12, 
               bbox=dict(facecolor='lightgray', alpha=0.5))
    
    # Mostrar el gr√°fico
    plt.tight_layout()
    plt.show(block=False)
    plt.pause(0.1)  # Peque√±a pausa para asegurar que se muestre


#==================================================
# 8. INTERFAZ DE USUARIO
#==================================================

# Iniciar una sesi√≥n de chat interactiva con el agente
def iniciar_chat_interactivo():
    logger.info("=== INICIANDO SESI√ìN DE CHAT INTERACTIVO ===")
    print("\n" + "="*50)
    print("ASISTENTE CONVERSACIONAL CON LANGRAPH")
    print("="*50)
    print("¬°Hola! Soy tu asistente conversacional con herramientas.")
    print("Puedo ayudarte con c√°lculos matem√°ticos y muchas otras cosas.")
    print("Escribe 'adi√≥s' o 'salir' para terminar.")
    print("Al final de cada interacci√≥n, ver√°s un diagrama del flujo de procesamiento.")
    print("-"*50)
    
    # Crear el agente
    logger.info("Construyendo el agente")
    agent = build_conversational_agent()
    logger.info("Agente construido exitosamente")
    
    # Inicializar el estado con el mensaje del sistema
    messages = [SystemMessage(content=SYSTEM_PROMPT)]
    logger.info("Estado inicial creado con mensaje de sistema")
    
    while True:
        # Limpiar el historial de ejecuci√≥n para esta nueva interacci√≥n
        global execution_path
        execution_path = []
        
        # Obtener entrada del usuario
        user_input = input("\nT√∫: ")
        logger.info(f"Input del usuario recibido: {user_input}")
        
        # Verificar si el usuario quiere terminar
        if user_input.lower() in ["adi√≥s", "adios", "chao", "hasta luego", "salir", "exit", "quit"]:
            logger.info("Usuario solicit√≥ terminar la conversaci√≥n")
            print("\nAsistente: ¬°Hasta luego! Ha sido un placer conversar contigo.")
            logger.info("=== SESI√ìN DE CHAT FINALIZADA ===")
            break
        
        # Registrar consulta del usuario en el historial de ejecuci√≥n
        execution_path.append({
            "node": "user_query",
            "description": f"Usuario pregunta: {user_input}"
        })
        
        # Agregar mensaje del usuario al historial
        messages.append(HumanMessage(content=user_input))
        logger.info("Mensaje del usuario a√±adido al historial")
        
        # Preparar el estado para esta iteraci√≥n
        state = {"messages": messages, "next_step": "decide_action"}
        logger.info(f"Estado preparado para invocar al agente: {format_state_for_log(state)}")
        
        # Invocar al agente
        logger.info("Invocando al agente")
        result = agent.invoke(state)
        logger.info("Agente invocado exitosamente")
        
        # Actualizar el historial de mensajes
        messages = result["messages"]
        logger.info(f"Historial de mensajes actualizado: {len(messages)} mensajes")
        
        # Obtener la respuesta final
        response = messages[-1].content
        
        # Registrar respuesta final en el historial de ejecuci√≥n
        execution_path.append({
            "node": "agent_response",
            "description": f"Respuesta final: {response[:50]}..." if len(response) > 50 else response
        })
        
        # Mostrar la respuesta
        print(f"\nAsistente: {response}")
        logger.info(f"Respuesta mostrada al usuario: {response[:50]}...")
        
        # Generar diagrama de flujo
        print("\nGenerando diagrama del flujo de procesamiento...")
        generar_diagrama_flujo(user_input, response)


#==================================================
# PUNTO DE ENTRADA PRINCIPAL
#==================================================

if __name__ == "__main__":
    logger.info("Iniciando programa")
    try:
        # Iniciar chat interactivo
        iniciar_chat_interactivo()
    except Exception as e:
        logger.error(f"Error en la ejecuci√≥n: {str(e)}", exc_info=True)
        print(f"Ocurri√≥ un error: {str(e)}")
    finally:
        # Cerrar cualquier figura abierta
        plt.close('all')
````

````python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
TDR Agente Multiagente con LangGraph y Ollama
=============================================

Este sistema implementa un agente conversacional multiagente para procesar 
T√©rminos de Referencia (TDR) y generar propuestas t√©cnicas profesionales.

Utiliza LangGraph para definir el flujo de trabajo entre diferentes agentes especializados
y Ollama con el modelo deepseek-r1:1.5b:32b para las tareas de procesamiento de lenguaje natural.

El sistema se integra con Telegram para permitir a los usuarios enviar archivos PDF
y recibir propuestas t√©cnicas generadas autom√°ticamente.
"""

import os
import re
import time
import json
import logging
import PyPDF2
import nest_asyncio
from typing import TypedDict, Annotated, List, Literal, Union, Dict, Any, Optional

# Telegram imports
from telegram import Update
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes

# LangGraph y LangChain imports
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage, FunctionMessage
from langchain_core.tools import tool
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langgraph.graph import StateGraph, END

# Imports para visualizaci√≥n
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib

# Para entornos Jupyter si es necesario
nest_asyncio.apply()

#==================================================
# 1. CONFIGURACI√ìN INICIAL
#==================================================

# Configuraci√≥n de logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.StreamHandler()
    ]
)
logger = logging.getLogger("TDR_Agente_LangGraph")

# Definir el estado para el sistema multiagente de an√°lisis de TDR
class TDRAgentState(TypedDict):
    messages: List[Annotated[Union[SystemMessage, HumanMessage, AIMessage, FunctionMessage], "Historial de la conversaci√≥n"]]
    next_step: Literal["extract_text", "analyze_tdr", "generate_index", "generate_sections", "combine_proposal", "evaluate_proposal", "end"]
    tdr_text: Optional[str]  # Texto extra√≠do del PDF
    tdr_info: Optional[str]  # Informaci√≥n extra√≠da del TDR
    index: Optional[Dict[str, str]]  # √çndice y descripciones
    sections: Optional[List[str]]  # Secciones generadas
    proposal: Optional[str]  # Propuesta final
    evaluation: Optional[str]  # Evaluaci√≥n de la propuesta
    pdf_path: Optional[str]  # Ruta al archivo PDF
    current_section_index: Optional[int]  # √çndice de la secci√≥n actual que se est√° generando

# Registro de la ruta tomada durante la ejecuci√≥n
execution_path = []

# Funci√≥n para formatear el estado para logs
def format_state_for_log(state):
    """Formatea el estado del agente para mostrar en logs"""
    messages = []
    for msg in state.get("messages", []):
        msg_type = type(msg).__name__
        if isinstance(msg, SystemMessage):
            content = msg.content[:50] + "..." if len(msg.content) > 50 else msg.content
            messages.append(f"[{msg_type}]: {content}")
        else:
            content = msg.content[:50] + "..." if len(msg.content) > 50 else msg.content
            messages.append(f"[{msg_type}]: {content}")
    
    formatted = {
        "message_count": len(messages),
        "message_types": messages,
        "next_step": state.get("next_step", None)
    }
    
    # A√±adir otros campos si existen
    for key in state:
        if key not in ["messages", "next_step"]:
            if key in ["tdr_text", "tdr_info", "proposal", "evaluation"] and state.get(key):
                # Truncar campos de texto largos
                formatted[key] = state[key][:100] + "..." if state[key] and len(state[key]) > 100 else state[key]
            else:
                formatted[key] = state[key]
            
    return formatted


#==================================================
# 2. CONFIGURACI√ìN DEL MODELO LLM
#==================================================

# Configurar Ollama con el modelo deepseek-r1:1.5b:32b
llm = Ollama(model="deepseek-r1:1.5b:32b")
logger.info(f"Modelo Ollama inicializado: deepseek-r1:1.5b:32b")

# Sistema de instrucciones para el agente
SYSTEM_PROMPT = """Eres un sistema multi-agente especializado en analizar T√©rminos de Referencia (TDR) y generar propuestas t√©cnicas profesionales.

El sistema est√° compuesto por varios agentes especializados que trabajan en conjunto:

1. Agente Extractor: Extrae texto de documentos PDF
2. Agente Analista: Analiza el TDR para extraer informaci√≥n clave como alcance, entregables, etc.
3. Agente Estructurador: Crea un √≠ndice para la propuesta t√©cnica basado en el an√°lisis
4. Agente Redactor: Genera contenido para cada secci√≥n de la propuesta
5. Agente Integrador: Combina todas las secciones en un documento cohesivo
6. Agente Evaluador: Verifica la calidad y adecuaci√≥n de la propuesta al TDR

Cada agente puede acceder a herramientas especializadas para cumplir su funci√≥n. El resultado final ser√° una propuesta t√©cnica completa que responde a los requisitos del TDR recibido.

Act√∫a de manera profesional, detallada y precisa en todas las etapas del proceso.
"""

logger.info("Sistema de prompt configurado")


#==================================================
# HERRAMIENTAS PARA EL AN√ÅLISIS DE TDR
#==================================================

def limpiar_texto(texto):
    """
    Limpia el texto eliminando espacios en blanco innecesarios y saltos de l√≠nea duplicados.
    """
    # Elimina espacios en blanco antes y despu√©s de cada salto de l√≠nea
    texto = re.sub(r'\s*\n\s*', '\n', texto)
    # Sustituye m√∫ltiples saltos de l√≠nea consecutivos por uno solo
    return re.sub(r'\n+', '\n', texto).strip()

@tool
def extract_text_from_pdf(pdf_file_path: str) -> str:
    """
    Extrae y concatena el texto de todas las p√°ginas del PDF.
    
    Args:
        pdf_file_path: Ruta al archivo PDF
        
    Returns:
        Texto extra√≠do del PDF
    """
    logger.info(f"Extrayendo texto del PDF: {pdf_file_path}")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "extract_text",
        "description": f"Extrayendo texto del PDF: {pdf_file_path}"
    })
    
    try:
        text_content = []
        with open(pdf_file_path, 'rb') as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text_content.append(page.extract_text() or "")
        
        text_content = "\n".join(text_content)
        cleaned_text = limpiar_texto(text_content)
        
        logger.info(f"Texto extra√≠do exitosamente: {len(cleaned_text)} caracteres")
        
        # Registrar √©xito
        execution_path.append({
            "node": "extract_text_result",
            "description": f"Texto extra√≠do: {len(cleaned_text)} caracteres"
        })
        
        return cleaned_text
    except Exception as e:
        error_message = f"Error al extraer texto: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "extract_text_error",
            "description": error_message
        })
        
        return f"Error: {error_message}"

@tool
def analyze_tdr(text: str) -> str:
    """
    Usa el modelo deepseek-r1:1.5b:32b para extraer informaci√≥n clave del TDR.
    
    Args:
        text: Texto del TDR
        
    Returns:
        Informaci√≥n extra√≠da del TDR en formato JSON
    """
    logger.info("Analizando el TDR")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "analyze_tdr",
        "description": "Extrayendo informaci√≥n clave del TDR"
    })
    
    try:
        prompt = (
            "Extrae la siguiente informaci√≥n del TDR (T√©rminos de Referencia):\n"
            "- Alcance del proyecto\n"
            "- Lista de entregables\n"
            "- Materiales y equipos requeridos\n"
            "- Lista de actividades\n"
            "- Normativas mencionadas\n\n"
            "Responde en formato JSON con claves en espa√±ol.\n\n"
            f"TDR:\n{text[:10000]}"  # Limitar a los primeros 10000 caracteres
        )
        
        response = llm.invoke(prompt)
        
        # Limpiar la respuesta si contiene delimitadores de c√≥digo
        if "```json" in response:
            response = response.split("```json")[1].split("```")[0].strip()
        elif "```" in response:
            response = response.split("```")[1].split("```")[0].strip()
        
        logger.info(f"An√°lisis del TDR completado: {response[:200]}...")
        
        # Registrar √©xito
        execution_path.append({
            "node": "analyze_tdr_result",
            "description": "An√°lisis completado exitosamente"
        })
        
        return response
    except Exception as e:
        error_message = f"Error al analizar el TDR: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "analyze_tdr_error",
            "description": error_message
        })
        
        return f"Error: {error_message}"

@tool
def generate_index(info: str) -> str:
    """
    Genera un √≠ndice para la propuesta t√©cnica basado en la informaci√≥n extra√≠da.
    
    Args:
        info: Informaci√≥n extra√≠da del TDR
        
    Returns:
        √çndice en formato JSON
    """
    logger.info("Generando √≠ndice para la propuesta t√©cnica")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "generate_index",
        "description": "Generando estructura e √≠ndice de la propuesta"
    })
    
    try:
        prompt = (
            "Con la siguiente informaci√≥n extra√≠da del TDR, genera un √≠ndice para la propuesta t√©cnica. "
            "Cada secci√≥n debe ser una clave y su valor, una breve descripci√≥n de lo que contendr√°. "
            "Incluye secciones como: Introducci√≥n, Objetivo, Alcance, Metodolog√≠a, Plan de Trabajo, "
            "Entregables, Recursos y Normativas. "
            "Devuelve la respuesta √∫nicamente en formato JSON sin ning√∫n texto adicional.\n\n"
            "Informaci√≥n:\n"
            f"{info}\n\n"
            "Responde en formato JSON."
        )
        
        response = llm.invoke(prompt)
        
        # Limpiar la respuesta JSON
        if "```json" in response:
            response = response.split("```json")[1].split("```")[0].strip()
        elif "```" in response:
            response = response.split("```")[1].split("```")[0].strip()
        
        # Validar que sea un JSON v√°lido
        try:
            json_response = json.loads(response)
            response = json.dumps(json_response)
        except json.JSONDecodeError:
            logger.warning("La respuesta no es un JSON v√°lido. Intentando reparar...")
            # Intentar reparar JSON
            response = re.sub(r',\s*}', '}', response)
            
        logger.info(f"√çndice generado: {response[:200]}...")
        
        # Registrar √©xito
        execution_path.append({
            "node": "generate_index_result",
            "description": "√çndice generado exitosamente"
        })
        
        return response
    except Exception as e:
        error_message = f"Error al generar √≠ndice: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "generate_index_error",
            "description": error_message
        })
        
        return f"Error: {error_message}"

@tool
def generate_section(params: str) -> str:
    """
    Redacta una secci√≥n espec√≠fica de la propuesta t√©cnica.
    
    Args:
        params: String en formato JSON con los siguientes campos:
               - section_name: Nombre de la secci√≥n
               - description: Descripci√≥n de la secci√≥n del √≠ndice
               - info: Informaci√≥n extra√≠da del TDR
        
    Returns:
        Texto de la secci√≥n generada
    """
    try:
        # Parsear los par√°metros JSON
        params_dict = json.loads(params)
        section_name = params_dict.get("section_name", "")
        description = params_dict.get("description", "")
        info = params_dict.get("info", "")
        
        logger.info(f"Generando secci√≥n: {section_name}")
        
        # Registrar en el historial de ejecuci√≥n
        execution_path.append({
            "node": "generate_section",
            "description": f"Redactando secci√≥n: {section_name}"
        })
        
        prompt = (
            f"Redacta la secci√≥n de '{section_name}' para la propuesta t√©cnica. "
            "Utiliza la siguiente descripci√≥n como gu√≠a:\n"
            f"{description}\n\n"
            "Adicionalmente, toma en cuenta la siguiente informaci√≥n extra√≠da del TDR:\n"
            f"{info}\n\n"
            "Responde con el texto de la secci√≥n en formato profesional y detallado. "
            "Usa lenguaje t√©cnico apropiado y estructura el contenido en p√°rrafos claros. "
            "La extensi√≥n debe ser proporcional a la importancia de la secci√≥n."
        )
        
        response = llm.invoke(prompt)
        
        logger.info(f"Secci√≥n '{section_name}' generada: {len(response)} caracteres")
        
        # Registrar √©xito
        execution_path.append({
            "node": "generate_section_result",
            "description": f"Secci√≥n {section_name} generada ({len(response)} caracteres)"
        })
        
        return f"## {section_name}\n\n{response}"
    except Exception as e:
        section_name = json.loads(params).get("section_name", "desconocida") if isinstance(params, str) else "desconocida"
        error_message = f"Error al generar secci√≥n {section_name}: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "generate_section_error",
            "description": error_message
        })
        
        return f"Error en secci√≥n {section_name}: {error_message}"

@tool
def combine_sections(params_str: str) -> str:
    """
    Combina todas las secciones en un solo documento de propuesta.
    
    Args:
        params_str: String en formato JSON con una lista de secciones
        
    Returns:
        Propuesta t√©cnica completa
    """
    logger.info("Combinando secciones en propuesta final")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "combine_sections",
        "description": "Integrando todas las secciones en un documento final"
    })
    
    try:
        # Parsear los par√°metros JSON
        params = json.loads(params_str)
        sections = params.get("sections", [])
        
        # A√±adir t√≠tulo principal
        proposal = "# PROPUESTA T√âCNICA\n\n"
        
        # Combinar secciones
        proposal += "\n\n".join(sections)
        
        logger.info(f"Propuesta combinada: {len(proposal)} caracteres")
        
        # Registrar √©xito
        execution_path.append({
            "node": "combine_sections_result",
            "description": f"Propuesta generada: {len(proposal)} caracteres"
        })
        
        return proposal
    except Exception as e:
        error_message = f"Error al combinar secciones: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "combine_sections_error",
            "description": error_message
        })
        
        return f"Error: {error_message}"

@tool
def evaluate_proposal(params_str: str) -> str:
    """
    Eval√∫a la propuesta final para validar que cumpla con lo solicitado.
    
    Args:
        params_str: String en formato JSON con:
                 - proposal: Texto de la propuesta t√©cnica
                 - tdr_info: Informaci√≥n extra√≠da del TDR
        
    Returns:
        Evaluaci√≥n en formato JSON
    """
    logger.info("Evaluando la propuesta t√©cnica")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "evaluate_proposal",
        "description": "Verificando calidad y alineaci√≥n con los requisitos"
    })
    
    try:
        # Parsear los par√°metros JSON
        params = json.loads(params_str)
        proposal = params.get("proposal", "")
        tdr_info = params.get("tdr_info", "")
        
        prompt = (
            "Eval√∫a la siguiente propuesta t√©cnica y verifica que cumpla con los requerimientos del TDR. "
            "Si detectas errores o √°reas de mejora, indica las correcciones necesarias.\n\n"
            "Informaci√≥n del TDR:\n"
            f"{tdr_info}\n\n"
            "Propuesta T√©cnica (parcial):\n"
            f"{proposal[:5000]}...\n\n"  # Limitamos a los primeros 5000 caracteres
            "Responde en formato JSON con los siguientes campos:\n"
            "1. status: 'aprobado', 'aprobado_con_cambios', o 'rechazado'\n"
            "2. puntuacion: valor num√©rico del 1 al 10\n"
            "3. fortalezas: lista de puntos fuertes\n"
            "4. debilidades: lista de puntos d√©biles\n"
            "5. recomendaciones: lista de mejoras sugeridas"
        )
        
        response = llm.invoke(prompt)
        
        # Limpiar la respuesta JSON
        if "```json" in response:
            response = response.split("```json")[1].split("```")[0].strip()
        elif "```" in response:
            response = response.split("```")[1].split("```")[0].strip()
        
        logger.info(f"Evaluaci√≥n completada: {response[:200]}...")
        
        # Registrar √©xito
        execution_path.append({
            "node": "evaluate_proposal_result",
            "description": "Evaluaci√≥n completada exitosamente"
        })
        
        return response
    except Exception as e:
        error_message = f"Error al evaluar propuesta: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "evaluate_proposal_error",
            "description": error_message
        })
        
        return f"Error: {error_message}"

@tool
def save_proposal_to_txt(proposal: str, filename: str = "propuesta.txt") -> str:
    """
    Guarda la propuesta final en un archivo TXT.
    
    Args:
        proposal: Texto de la propuesta t√©cnica
        filename: Nombre del archivo de salida
        
    Returns:
        Ruta del archivo guardado
    """
    logger.info(f"Guardando propuesta en archivo: {filename}")
    
    # Registrar en el historial de ejecuci√≥n
    execution_path.append({
        "node": "save_proposal",
        "description": f"Guardando en archivo: {filename}"
    })
    
    try:
        with open(filename, "w", encoding="utf-8") as f:
            f.write(proposal)
        
        logger.info(f"Propuesta guardada exitosamente: {filename}")
        
        # Registrar √©xito
        execution_path.append({
            "node": "save_proposal_result",
            "description": f"Archivo guardado: {filename}"
        })
        
        return filename
    except Exception as e:
        error_message = f"Error al guardar propuesta: {str(e)}"
        logger.error(error_message)
        
        # Registrar error
        execution_path.append({
            "node": "save_proposal_error",
            "description": error_message
        })
        
        return f"Error: {error_message}"


#==================================================
# 4. NODOS DEL GRAFO PARA AN√ÅLISIS DE TDR
#==================================================

# Nodo para extraer texto del PDF
def extract_text_node(state: TDRAgentState) -> TDRAgentState:
    """Extrae texto del PDF del TDR"""
    logger.info(f"Iniciando extract_text_node con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "extract_text_node",
        "description": "Nodo de extracci√≥n de texto del PDF"
    })
    
    # Verificar que exista la ruta al PDF
    pdf_path = state.get("pdf_path")
    if not pdf_path:
        logger.error("No se proporcion√≥ la ruta del PDF")
        state["messages"].append(AIMessage(content="Error: No se proporcion√≥ la ruta del PDF"))
        state["next_step"] = "end"
        return state
    
    # Extraer texto del PDF
    text = extract_text_from_pdf(pdf_path)
    if text.startswith("Error:"):
        logger.error(f"Error al extraer texto: {text}")
        state["messages"].append(AIMessage(content=text))
        state["next_step"] = "end"
        return state
    
    # Guardar el texto extra√≠do en el estado
    state["tdr_text"] = text
    logger.info(f"Texto extra√≠do: {len(text)} caracteres")
    
    # Actualizar el mensaje en el historial
    state["messages"].append(AIMessage(content=f"Texto extra√≠do exitosamente: {len(text)} caracteres"))
    
    # Siguiente paso
    state["next_step"] = "analyze_tdr"
    
    logger.info(f"Estado despu√©s de extract_text_node: {format_state_for_log(state)}")
    return state

# Nodo para analizar el TDR
def analyze_tdr_node(state: TDRAgentState) -> TDRAgentState:
    """Analiza el TDR para extraer informaci√≥n clave"""
    logger.info(f"Iniciando analyze_tdr_node con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "analyze_tdr_node",
        "description": "Nodo de an√°lisis del TDR"
    })
    
    # Verificar que exista el texto del TDR
    tdr_text = state.get("tdr_text")
    if not tdr_text:
        logger.error("No hay texto del TDR para analizar")
        state["messages"].append(AIMessage(content="Error: No hay texto del TDR para analizar"))
        state["next_step"] = "end"
        return state
    
    # Analizar el TDR
    tdr_info = analyze_tdr(tdr_text)
    if tdr_info.startswith("Error:"):
        logger.error(f"Error al analizar TDR: {tdr_info}")
        state["messages"].append(AIMessage(content=tdr_info))
        state["next_step"] = "end"
        return state
    
    # Guardar la informaci√≥n extra√≠da en el estado
    state["tdr_info"] = tdr_info
    logger.info(f"TDR analizado: {len(tdr_info)} caracteres")
    
    # Actualizar el mensaje en el historial
    state["messages"].append(AIMessage(content=f"TDR analizado exitosamente. Informaci√≥n extra√≠da: {tdr_info[:200]}..."))
    
    # Siguiente paso
    state["next_step"] = "generate_index"
    
    logger.info(f"Estado despu√©s de analyze_tdr_node: {format_state_for_log(state)}")
    return state

# Nodo para generar el √≠ndice de la propuesta
def generate_index_node(state: TDRAgentState) -> TDRAgentState:
    """Genera el √≠ndice para la propuesta t√©cnica"""
    logger.info(f"Iniciando generate_index_node con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "generate_index_node",
        "description": "Nodo de generaci√≥n del √≠ndice"
    })
    
    # Verificar que exista la informaci√≥n del TDR
    tdr_info = state.get("tdr_info")
    if not tdr_info:
        logger.error("No hay informaci√≥n del TDR para generar el √≠ndice")
        state["messages"].append(AIMessage(content="Error: No hay informaci√≥n del TDR para generar el √≠ndice"))
        state["next_step"] = "end"
        return state
    
    # Generar el √≠ndice
    index_str = generate_index(tdr_info)
    if index_str.startswith("Error:"):
        logger.error(f"Error al generar √≠ndice: {index_str}")
        state["messages"].append(AIMessage(content=index_str))
        state["next_step"] = "end"
        return state
    
    try:
        # Convertir el √≠ndice a diccionario
        index_dict = json.loads(index_str)
        
        # Guardar el √≠ndice en el estado
        state["index"] = index_dict
        state["sections"] = []  # Inicializar lista de secciones
        state["current_section_index"] = 0  # Inicializar √≠ndice para generar secciones
        
        # Crear una lista con los nombres de las secciones
        section_names = list(index_dict.keys())
        
        logger.info(f"√çndice generado con {len(section_names)} secciones")
        
        # Actualizar el mensaje en el historial
        index_summary = ", ".join(section_names[:5])
        if len(section_names) > 5:
            index_summary += f", y {len(section_names) - 5} m√°s"
        
        state["messages"].append(AIMessage(content=f"√çndice generado exitosamente con las siguientes secciones: {index_summary}"))
        
        # Siguiente paso
        state["next_step"] = "generate_sections"
        
    except json.JSONDecodeError as e:
        error_message = f"Error al decodificar el √≠ndice JSON: {str(e)}, √≠ndice: {index_str[:200]}..."
        logger.error(error_message)
        state["messages"].append(AIMessage(content=f"Error: {error_message}"))
        state["next_step"] = "end"
    
    logger.info(f"Estado despu√©s de generate_index_node: {format_state_for_log(state)}")
    return state

# Nodo para generar secciones de la propuesta
def generate_sections_node(state: TDRAgentState) -> TDRAgentState:
    """Genera una secci√≥n de la propuesta t√©cnica a la vez"""
    logger.info(f"Iniciando generate_sections_node con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "generate_sections_node",
        "description": "Nodo de generaci√≥n de secciones"
    })
    
    # Verificar que existan el √≠ndice y la informaci√≥n del TDR
    index = state.get("index")
    tdr_info = state.get("tdr_info")
    current_index = state.get("current_section_index", 0)
    
    if not index or not tdr_info:
        logger.error("Faltan datos necesarios para generar secciones")
        state["messages"].append(AIMessage(content="Error: Faltan datos necesarios para generar secciones"))
        state["next_step"] = "end"
        return state
    
    # Obtener lista de secciones
    section_names = list(index.keys())
    
    # Verificar si ya se procesaron todas las secciones
    if current_index >= len(section_names):
        logger.info("Todas las secciones han sido generadas")
        state["messages"].append(AIMessage(content="Todas las secciones han sido generadas exitosamente"))
        state["next_step"] = "combine_proposal"
        return state
    
    # Obtener la secci√≥n actual
    section_name = section_names[current_index]
    section_description = index[section_name]
    
    logger.info(f"Generando secci√≥n {current_index+1}/{len(section_names)}: {section_name}")
    
    # Generar la secci√≥n - pasar par√°metros como un solo JSON
    params = json.dumps({
        "section_name": section_name,
        "description": section_description,
        "info": tdr_info
    })
    section_content = generate_section(params)
    if section_content.startswith("Error:"):
        logger.error(f"Error al generar secci√≥n {section_name}: {section_content}")
        state["messages"].append(AIMessage(content=section_content))
        # Continuamos con la siguiente secci√≥n en caso de error
    else:
        # A√±adir la secci√≥n generada a la lista
        sections = state.get("sections", [])
        sections.append(section_content)
        state["sections"] = sections
        
        logger.info(f"Secci√≥n {section_name} generada: {len(section_content)} caracteres")
    
    # Actualizar mensaje en el historial
    state["messages"].append(AIMessage(content=f"Secci√≥n {current_index+1}/{len(section_names)} ({section_name}) generada"))
    
    # Incrementar el √≠ndice para la pr√≥xima secci√≥n
    state["current_section_index"] = current_index + 1
    
    # Mantener el mismo paso para generar la siguiente secci√≥n
    state["next_step"] = "generate_sections"
    
    logger.info(f"Estado despu√©s de generate_sections_node: {format_state_for_log(state)}")
    return state

# Nodo para combinar la propuesta completa
def combine_proposal_node(state: TDRAgentState) -> TDRAgentState:
    """Combina todas las secciones en una propuesta completa"""
    logger.info(f"Iniciando combine_proposal_node con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "combine_proposal_node",
        "description": "Nodo de integraci√≥n de la propuesta"
    })
    
    # Verificar que existan secciones generadas
    sections = state.get("sections", [])
    if not sections:
        logger.error("No hay secciones para combinar")
        state["messages"].append(AIMessage(content="Error: No hay secciones para combinar en una propuesta"))
        state["next_step"] = "end"
        return state
    
    # Combinar las secciones con formato JSON
    params = json.dumps({"sections": sections})
    proposal = combine_sections(params)
    if proposal.startswith("Error:"):
        logger.error(f"Error al combinar propuesta: {proposal}")
        state["messages"].append(AIMessage(content=proposal))
        state["next_step"] = "end"
        return state
    
    # Guardar la propuesta en el estado
    state["proposal"] = proposal
    logger.info(f"Propuesta combinada: {len(proposal)} caracteres")
    
    # Guardar la propuesta en un archivo
    filename = save_proposal_to_txt(proposal)
    if filename.startswith("Error:"):
        logger.warning(f"Error al guardar propuesta: {filename}")
        # Continuamos a pesar del error de guardado
    else:
        logger.info(f"Propuesta guardada en: {filename}")
        state["proposal_filename"] = filename
    
    # Actualizar mensaje en el historial
    state["messages"].append(AIMessage(content=f"Propuesta t√©cnica generada y combinada exitosamente con {len(sections)} secciones"))
    
    # Siguiente paso
    state["next_step"] = "evaluate_proposal"
    
    logger.info(f"Estado despu√©s de combine_proposal_node: {format_state_for_log(state)}")
    return state

# Nodo para evaluar la propuesta final
def evaluate_proposal_node(state: TDRAgentState) -> TDRAgentState:
    """Eval√∫a la calidad y adecuaci√≥n de la propuesta t√©cnica"""
    logger.info(f"Iniciando evaluate_proposal_node con estado: {format_state_for_log(state)}")
    
    # Registrar inicio del nodo
    execution_path.append({
        "node": "evaluate_proposal_node",
        "description": "Nodo de evaluaci√≥n de la propuesta"
    })
    
    # Verificar que existan la propuesta y la informaci√≥n del TDR
    proposal = state.get("proposal")
    tdr_info = state.get("tdr_info")
    
    if not proposal or not tdr_info:
        logger.error("Faltan datos necesarios para evaluar la propuesta")
        state["messages"].append(AIMessage(content="Error: Faltan datos necesarios para evaluar la propuesta"))
        state["next_step"] = "end"
        return state
    
    # Evaluar la propuesta - pasar par√°metros como un solo JSON
    params = json.dumps({
        "proposal": proposal,
        "tdr_info": tdr_info
    })
    evaluation = evaluate_proposal(params)
    if evaluation.startswith("Error:"):
        logger.error(f"Error al evaluar propuesta: {evaluation}")
        state["messages"].append(AIMessage(content=evaluation))
        # Continuamos a pesar del error de evaluaci√≥n
    else:
        # Guardar la evaluaci√≥n en el estado
        state["evaluation"] = evaluation
        logger.info(f"Propuesta evaluada: {evaluation[:200]}...")
        
    try:
        # Intentar convertir la evaluaci√≥n a diccionario para presentarla mejor
        eval_dict = json.loads(evaluation)
        status = eval_dict.get("status", "desconocido")
        puntuacion = eval_dict.get("puntuacion", "N/A")
        
        # Actualizar mensaje en el historial con resumen de la evaluaci√≥n
        state["messages"].append(AIMessage(
            content=f"Propuesta evaluada: Estado '{status}' con puntuaci√≥n {puntuacion}/10. "
                   f"La propuesta ha sido completada y guardada en el archivo '{state.get('proposal_filename', 'propuesta.txt')}'."
        ))
    except (json.JSONDecodeError, TypeError):
        # Si hay error al decodificar el JSON, usar texto plano
        state["messages"].append(AIMessage(
            content=f"Evaluaci√≥n completada. La propuesta ha sido guardada en el archivo '{state.get('proposal_filename', 'propuesta.txt')}'."
        ))
    
    # Siguiente paso (fin del proceso)
    state["next_step"] = "end"
    
    logger.info(f"Estado despu√©s de evaluate_proposal_node: {format_state_for_log(state)}")
    return state

# Funci√≥n para decidir el siguiente paso
def router(state: TDRAgentState) -> Literal["extract_text", "analyze_tdr", "generate_index", "generate_sections", "combine_proposal", "evaluate_proposal", "end"]:
    """Determina el siguiente paso en el flujo de trabajo"""
    next_step = state["next_step"]
    logger.info(f"Router decidiendo siguiente paso: {next_step}")
    
    # Registrar decisi√≥n del router
    execution_path.append({
        "node": "router",
        "description": f"Siguiente paso: {next_step}"
    })
    
    return next_step


#==================================================
# 5. CONSTRUCCI√ìN DEL GRAFO
#==================================================

# Crear el grafo del agente para an√°lisis de TDR
def build_tdr_agent():
    logger.info("Construyendo el grafo del agente para an√°lisis de TDR")
    
    # Inicializar el grafo con el tipo de estado
    workflow = StateGraph(TDRAgentState)
    
    # Agregar nodos
    workflow.add_node("extract_text", extract_text_node)
    workflow.add_node("analyze_tdr", analyze_tdr_node)
    workflow.add_node("generate_index", generate_index_node)
    workflow.add_node("generate_sections", generate_sections_node)
    workflow.add_node("combine_proposal", combine_proposal_node)
    workflow.add_node("evaluate_proposal", evaluate_proposal_node)
    logger.info("Nodos a√±adidos al grafo")
    
    # Establecer el nodo de inicio
    workflow.set_entry_point("extract_text")
    logger.info("Punto de entrada establecido: extract_text")
    
    # Conectar los nodos seg√∫n la decisi√≥n del router
    workflow.add_conditional_edges("extract_text", router, {
        "analyze_tdr": "analyze_tdr",
        "end": END
    })
    
    workflow.add_conditional_edges("analyze_tdr", router, {
        "generate_index": "generate_index",
        "end": END
    })
    
    workflow.add_conditional_edges("generate_index", router, {
        "generate_sections": "generate_sections",
        "end": END
    })
    
    workflow.add_conditional_edges("generate_sections", router, {
        "generate_sections": "generate_sections",  # Loop para generar todas las secciones
        "combine_proposal": "combine_proposal",
        "end": END
    })
    
    workflow.add_conditional_edges("combine_proposal", router, {
        "evaluate_proposal": "evaluate_proposal",
        "end": END
    })
    
    workflow.add_conditional_edges("evaluate_proposal", router, {
        "end": END
    })
    
    logger.info("Bordes condicionales a√±adidos")
    
    # Compilar el grafo
    logger.info("Compilando el grafo")
    return workflow.compile()


#==================================================
# 6. VISUALIZACI√ìN DEL FLUJO DE TRABAJO
#==================================================

def generar_diagrama_flujo():
    """
    Genera un diagrama del flujo del agente basado en la ruta de ejecuci√≥n.
    """
    logger.info("Generando diagrama del flujo de trabajo")
    
    # Crear un nuevo grafo para la visualizaci√≥n
    G = nx.DiGraph()
    
    # A√±adir nodos principales del agente
    G.add_node("extract_text", pos=(0, 0), node_type="extraction", label="Extracci√≥n\nde Texto")
    G.add_node("analyze_tdr", pos=(2, 0), node_type="analysis", label="An√°lisis\ndel TDR")
    G.add_node("generate_index", pos=(4, 0), node_type="generation", label="Generaci√≥n\nde √çndice")
    G.add_node("generate_sections", pos=(6, 0), node_type="generation", label="Generaci√≥n\nde Secciones")
    G.add_node("combine_proposal", pos=(8, 0), node_type="integration", label="Integraci√≥n\nde Propuesta")
    G.add_node("evaluate_proposal", pos=(10, 0), node_type="evaluation", label="Evaluaci√≥n\nde Propuesta")
    
    # A√±adir aristas b√°sicas
    G.add_edge("extract_text", "analyze_tdr", label="Texto\nextra√≠do")
    G.add_edge("analyze_tdr", "generate_index", label="TDR\nanalizado")
    G.add_edge("generate_index", "generate_sections", label="√çndice\ngenerado")
    G.add_edge("generate_sections", "generate_sections", label="Siguiente\nsecci√≥n")
    G.add_edge("generate_sections", "combine_proposal", label="Todas las\nsecciones")
    G.add_edge("combine_proposal", "evaluate_proposal", label="Propuesta\nintegrada")
    
    # Crear figura
    plt.figure(figsize=(16, 8))
    
    # Posiciones de los nodos
    pos = nx.get_node_attributes(G, 'pos')
    
    # Colores para los nodos seg√∫n su tipo
    node_colors = {
        'extraction': '#66c2a5',    # Verde azulado para extracci√≥n
        'analysis': '#fc8d62',      # Naranja para an√°lisis
        'generation': '#8da0cb',    # Azul para generaci√≥n
        'integration': '#e78ac3',   # Rosa para integraci√≥n
        'evaluation': '#a6d854'     # Verde para evaluaci√≥n
    }
    
    # Preparar colores para cada nodo
    color_map = [node_colors[G.nodes[node].get('node_type', 'extraction')] for node in G.nodes()]
    
    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, node_size=3000, node_color=color_map, alpha=0.8, 
                          node_shape='o', edgecolors='black', linewidths=2)
    
    # Preparar etiquetas de nodos (con formato)
    node_labels = {}
    for node in G.nodes():
        label = G.nodes[node].get('label', node)
        node_labels[node] = label
    
    # Dibujar etiquetas de nodos
    nx.draw_networkx_labels(G, pos, node_labels, font_size=10, font_weight='bold')
    
    # Dibujar todas las aristas
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2.0, 
                          edge_color='gray', style='solid', 
                          arrows=True, arrowsize=15, 
                          connectionstyle='arc3,rad=0.1')
    
    # Etiquetas de aristas
    edge_labels = {}
    for u, v, data in G.edges(data=True):
        if 'label' in data:
            edge_labels[(u, v)] = data['label']
    
    # Dibujar etiquetas de aristas
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, 
                               font_weight='bold', label_pos=0.5, rotate=False)
    
    # A√±adir t√≠tulo
    plt.title("Flujo de Trabajo del Sistema Multiagente para An√°lisis de TDR", fontsize=16, fontweight='bold')
    
    # Quitar ejes
    plt.axis('off')
    
    # Mostrar el gr√°fico
    plt.tight_layout()
    plt.savefig("flujo_tdr.png")
    logger.info("Diagrama guardado como flujo_tdr.png")
    plt.close()
    
    return "flujo_tdr.png"


#==================================================
# 7. BOT DE TELEGRAM
#==================================================

# Configuraci√≥n del token del bot
TELEGRAM_BOT_TOKEN = "8195110973:AAHEkkFWa0LcHHMwmhN3-Hxj9ZzbH4q2ptk"

# Handlers para Telegram
async def start(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    """Mensaje de bienvenida del bot"""
    await update.message.reply_text(
        "¬°Bienvenido al Bot de An√°lisis de TDR y Generaci√≥n de Propuestas T√©cnicas!\n\n"
        "Este bot utiliza un sistema multiagente con LangGraph y Ollama para procesar T√©rminos de Referencia "
        "y generar propuestas t√©cnicas profesionales.\n\n"
        "Para comenzar, env√≠a un archivo PDF con el TDR que deseas analizar."
    )

async def handle_document(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    """Procesa el archivo PDF recibido y ejecuta el flujo de trabajo"""
    # Verificar que sea un PDF
    document = update.message.document
    if document.mime_type != "application/pdf":
        await update.message.reply_text("Por favor, env√≠a un archivo PDF.")
        return
    
    # Informar al usuario que se inicia el procesamiento
    await update.message.reply_text("üìÑ PDF recibido. Iniciando el an√°lisis...")
    
    try:
        # Limpiar el historial de ejecuci√≥n
        global execution_path
        execution_path = []
        
        # Descargar el PDF
        file = await document.get_file()
        file_path = f"tdr_{int(time.time())}.pdf"
        pdf_file = await file.download_to_drive(custom_path=file_path)
        logger.info(f"PDF recibido y guardado en: {file_path}")
        
        # Crear el agente
        agent = build_tdr_agent()
        logger.info("Agente TDR construido exitosamente")
        
        # Inicializar el estado con mensajes y la ruta del PDF
        state = {
            "messages": [SystemMessage(content=SYSTEM_PROMPT)],
            "next_step": "extract_text",
            "pdf_path": file_path
        }
        
        # Informar al usuario que se est√° extrayendo el texto
        await update.message.reply_text("üîç Extrayendo texto del PDF...")
        
        # Invocar al agente para comenzar el procesamiento
        result = agent.invoke(state)
        logger.info("Procesamiento del TDR completado")
        
        # Obtener la propuesta final y el archivo generado
        proposal_filename = result.get("proposal_filename", "propuesta.txt")
        
        # Informar al usuario sobre el resultado
        if "evaluation" in result:
            try:
                eval_dict = json.loads(result["evaluation"])
                status = eval_dict.get("status", "completado")
                puntuacion = eval_dict.get("puntuacion", "N/A")
                await update.message.reply_text(
                    f"‚úÖ Propuesta generada y evaluada:\n"
                    f"‚Ä¢ Estado: {status}\n"
                    f"‚Ä¢ Puntuaci√≥n: {puntuacion}/10\n"
                    f"‚Ä¢ Archivo: {proposal_filename}"
                )
            except:
                await update.message.reply_text(f"‚úÖ Propuesta generada y guardada en: {proposal_filename}")
        else:
            await update.message.reply_text(f"‚úÖ Propuesta generada y guardada en: {proposal_filename}")
        
        # Enviar el archivo de la propuesta
        try:
            with open(proposal_filename, "rb") as f:
                await update.message.reply_document(document=f, filename=f"Propuesta_{int(time.time())}.txt")
            logger.info(f"Archivo enviado: {proposal_filename}")
        except Exception as e:
            logger.error(f"Error al enviar archivo: {str(e)}")
            await update.message.reply_text(f"‚ùå Error al enviar el archivo: {str(e)}")
        
        # Generar y enviar el diagrama del flujo
        try:
            diagram_path = generar_diagrama_flujo()
            with open(diagram_path, "rb") as f:
                await update.message.reply_photo(photo=f, caption="Diagrama del flujo de trabajo del sistema multiagente")
            logger.info(f"Diagrama enviado: {diagram_path}")
        except Exception as e:
            logger.error(f"Error al generar o enviar diagrama: {str(e)}")
        
    except Exception as e:
        logger.error(f"Error en el procesamiento: {str(e)}", exc_info=True)
        await update.message.reply_text(f"‚ùå Error en el procesamiento: {str(e)}")

# Funci√≥n principal para iniciar el bot
def main():
    """Iniciar el bot de Telegram"""
    logger.info("Iniciando bot de Telegram")
    
    # Crear la aplicaci√≥n
    application = Application.builder().token(TELEGRAM_BOT_TOKEN).build()
    
    # A√±adir handlers
    application.add_handler(CommandHandler("start", start))
    application.add_handler(MessageHandler(filters.Document.PDF, handle_document))
    
    # Iniciar polling
    application.run_polling()
    logger.info("Bot iniciado. Esperando mensajes...")


#==================================================
# PUNTO DE ENTRADA PRINCIPAL
#==================================================

if __name__ == "__main__":
    try:
        # Iniciar el bot de Telegram
        main()
    except Exception as e:
        logger.error(f"Error al iniciar el bot: {str(e)}", exc_info=True)
        print(f"Error: {str(e)}")
````

````python
import os
import json
import pickle
from typing import Dict, List, Optional
from collections import defaultdict

# Importaciones para documentos y embeddings
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Configuramos la ruta de la carpeta con PDFs y de salida
PDF_FOLDER_PATH = "Propuestas/"  # Carpeta con PDFs
INDEX_FOLDER_PATH = "indices/"    # Carpeta para guardar √≠ndices y metadatos
METADATA_FILE = "documentos_metadata.json"  # Archivo para guardar metadatos de documentos

# Modelo de embeddings a usar (modificado para usar uno est√°ndar)
MODELOS_EMBEDDINGS = [
    "nomic-embed-text:latest",    # Poniendo este primero con el nombre exacto
    "mxbai-embed-large:latest",   # Segundo con nombre exacto
    "tazarov/all-minilm-l6-v2-f32:latest"  # Tercero con nombre exacto
]

# Crear carpeta para √≠ndices si no existe
if not os.path.exists(INDEX_FOLDER_PATH):
    os.makedirs(INDEX_FOLDER_PATH)
    print(f"Creada carpeta {INDEX_FOLDER_PATH} para guardar √≠ndices y metadatos")

# Funci√≥n para cargar PDFs desde una carpeta
def load_pdfs_from_directory(directory_path):
    # Verificar si la carpeta existe
    if not os.path.exists(directory_path):
        print(f"Error: La carpeta {directory_path} no existe.")
        os.makedirs(directory_path)
        print(f"Se ha creado la carpeta {directory_path}. Por favor, a√±ade los PDFs y vuelve a ejecutar.")
        return [], {}
    
    # Comprobar si hay PDFs
    pdfs = [f for f in os.listdir(directory_path) if f.endswith('.pdf')]
    if not pdfs:
        print(f"Error: No se encontraron archivos PDF en {directory_path}.")
        print("Por favor, a√±ade algunos PDFs a la carpeta.")
        return [], {}
    
    # Estructura para almacenar metadatos de documentos completos
    documentos_metadata = {}
    all_documents = []
    
    # Cargar cada PDF individualmente para mantener referencia al documento completo
    for pdf_file in pdfs:
        pdf_path = os.path.join(directory_path, pdf_file)
        print(f"Cargando {pdf_file}...")
        
        try:
            # Cargar el documento
            loader = PyPDFLoader(pdf_path)
            pages = loader.load()
            
            # Almacenar metadatos del documento
            documentos_metadata[pdf_path] = {
                "nombre": pdf_file,
                "ruta": pdf_path,
                "paginas": len(pages),
                "indices_paginas": [i for i in range(len(pages))]
            }
            
            # A√±adir a la lista completa de documentos
            all_documents.extend(pages)
            print(f"  -> {len(pages)} p√°ginas cargadas correctamente")
            
        except Exception as e:
            print(f"  -> Error al cargar {pdf_file}: {str(e)}")
    
    print(f"Cargados {len(all_documents)} fragmentos de documentos PDF de {len(pdfs)} archivos.")
    return all_documents, documentos_metadata

# Funci√≥n para verificar modelos de embeddings disponibles
def verificar_modelos_embeddings():
    """Verifica qu√© modelos de embeddings est√°n disponibles y retorna el primero que funcione."""
    from langchain_community.llms import Ollama
    import requests
    
    print("Verificando modelos de embeddings disponibles en Ollama...")
    
    # Primero comprobar que Ollama est√° en ejecuci√≥n
    try:
        response = requests.get("http://localhost:11434/api/tags")
        if response.status_code != 200:
            print("‚ùå Error al conectar con Ollama. Aseg√∫rate de que el servicio est√© en ejecuci√≥n.")
            return None
            
        # Obtener modelos disponibles
        modelos_disponibles = [model["name"] for model in response.json()["models"]]
        print(f"Modelos disponibles en Ollama: {', '.join(modelos_disponibles)}")
        
        # Probar los modelos de embeddings en orden
        for modelo in MODELOS_EMBEDDINGS:
            if modelo in modelos_disponibles:
                print(f"Probando modelo de embeddings: {modelo}")
                try:
                    # Intenta crear embeddings con este modelo
                    embeddings = OllamaEmbeddings(
                        model=modelo,
                        base_url="http://localhost:11434",
                    )
                    # Probar el modelo con un texto simple
                    test_embed = embeddings.embed_query("Prueba de embeddings")
                    if len(test_embed) > 0:
                        print(f"‚úÖ Modelo de embeddings {modelo} funciona correctamente")
                        return modelo
                except Exception as e:
                    print(f"‚ùå Error al probar modelo {modelo}: {str(e)}")
        
        # Si llegamos aqu√≠, ning√∫n modelo funcion√≥
        print("‚ùå Ninguno de los modelos de embeddings conocidos est√° disponible")
        print("Descarga alguno de estos modelos o usa embeddings alternativos:")
        for modelo in MODELOS_EMBEDDINGS:
            print(f"  ollama pull {modelo}")
        return None
        
    except Exception as e:
        print(f"‚ùå Error al verificar Ollama: {str(e)}")
        return None

# Funci√≥n principal para crear el √≠ndice vectorial
def crear_indice_vectorial():
    print("=== CREACI√ìN DE √çNDICE VECTORIAL ===")
    print("Cargando documentos PDF...")
    
    # Cargar los documentos y sus metadatos
    docs_list, documentos_metadata = load_pdfs_from_directory(PDF_FOLDER_PATH)
    
    # Si no hay documentos, no seguimos
    if not docs_list:
        print("No se encontraron documentos. Por favor, a√±ade PDFs a la carpeta y vuelve a ejecutar.")
        return False
    
    # Guardar metadatos de documentos
    with open(os.path.join(INDEX_FOLDER_PATH, METADATA_FILE), 'w', encoding='utf-8') as f:
        json.dump(documentos_metadata, f, ensure_ascii=False, indent=2)
    print(f"Metadatos de documentos guardados en {os.path.join(INDEX_FOLDER_PATH, METADATA_FILE)}")
    
    print("Dividiendo documentos...")
    # Dividir los documentos en fragmentos m√°s manejables
    text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
        chunk_size=500,  # Fragmentos m√°s grandes para mejor contexto
        chunk_overlap=50
    )
    doc_splits = text_splitter.split_documents(docs_list)
    print(f"Documentos divididos en {len(doc_splits)} fragmentos m√°s peque√±os.")
    
    # Verificar modelos de embeddings disponibles
    modelo_embeddings = verificar_modelos_embeddings()
    if not modelo_embeddings:
        print("‚ùå No se encontr√≥ un modelo de embeddings v√°lido.")
        print("Opciones:")
        print("1. Instala alguno de los modelos de embeddings usando 'ollama pull [modelo]'")
        print("2. Modifica el c√≥digo para usar embeddings alternativos, por ejemplo HuggingFace o SentenceTransformers")
        return False
    
    print(f"Usando modelo de embeddings: {modelo_embeddings}")
    try:
        # Crear base de vectores con embeddings de Ollama
        embeddings = OllamaEmbeddings(
            model=modelo_embeddings,
            base_url="http://localhost:11434",
        )
        
        print("Creando base de vectores...")
        # Crear la base de vectores con FAISS
        vectorstore = FAISS.from_documents(
            documents=doc_splits,
            embedding=embeddings,
        )
        
        # Guardar la base de vectores
        index_path = os.path.join(INDEX_FOLDER_PATH, "faiss_index")
        vectorstore.save_local(index_path)
        print(f"Base de vectores guardada en {index_path}")
        
        # Guardar tambi√©n la lista completa de documentos para referencia futura
        documentos_path = os.path.join(INDEX_FOLDER_PATH, "documentos_completos.pkl")
        with open(documentos_path, 'wb') as f:
            pickle.dump(docs_list, f)
        print(f"Lista de documentos guardada en {documentos_path}")
        
        # Guardar el nombre del modelo de embeddings para referencia futura
        with open(os.path.join(INDEX_FOLDER_PATH, "modelo_embeddings.txt"), 'w') as f:
            f.write(modelo_embeddings)
        
        return True
        
    except Exception as e:
        print(f"Error al crear la base de vectores: {str(e)}")
        return False

# Funci√≥n principal
if __name__ == "__main__":
    print("\n" + "="*50)
    print("CREADOR DE √çNDICE VECTORIAL PARA DOCUMENTOS PDF")
    print("="*50 + "\n")
    
    # Crear el √≠ndice
    exito = crear_indice_vectorial()
    
    if exito:
        print("\n" + "="*50)
        print("‚úÖ √çNDICE VECTORIAL CREADO CON √âXITO")
        print("="*50)
        print("\nAhora puedes usar el agente de consulta para hacer preguntas sobre tus documentos.")
    else:
        print("\n" + "="*50)
        print("‚ùå ERROR AL CREAR EL √çNDICE VECTORIAL")
        print("="*50)
````

````python
import os
import json
import uuid
import pickle
import traceback
from typing import Annotated, Sequence, List, Dict, Any
from typing_extensions import TypedDict

# Importaciones para documentos y embeddings
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.llms import Ollama
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document

# Importaciones para LangGraph y herramientas
from langchain.tools.retriever import create_retriever_tool
from langgraph.graph.message import add_messages
from langgraph.graph import END, StateGraph, START

# Importaciones de LangChain para mensajes y modelos
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage, ToolMessage
from langchain_core.tools import tool

# Importaci√≥n para modelos de datos
from pydantic import BaseModel, Field

# Configuraci√≥n de rutas
INDEX_FOLDER_PATH = "indices/"    # Carpeta donde est√°n guardados los √≠ndices
FAISS_INDEX_PATH = os.path.join(INDEX_FOLDER_PATH, "faiss_index")
METADATA_FILE = os.path.join(INDEX_FOLDER_PATH, "documentos_metadata.json")
DOCUMENTOS_FILE = os.path.join(INDEX_FOLDER_PATH, "documentos_completos.pkl")
MODELO_EMBEDDINGS_FILE = os.path.join(INDEX_FOLDER_PATH, "modelo_embeddings.txt")

# Modelo Ollama a usar para LLM (aseg√∫rate de que el nombre coincida exactamente)
MODELO_OLLAMA = "deepseek-r1:32b"  # Nombre exacto como aparece en 'ollama list'

# Posibles modelos de embeddings a usar - nombres exactos como aparecen en 'ollama list'
MODELOS_EMBEDDINGS = [
    "nomic-embed-text:latest",
    "mxbai-embed-large:latest",
    "tazarov/all-minilm-l6-v2-f32:latest"
]

# Variable global para almacenar documentos completos
documentos_completos = {}

# Funci√≥n para obtener modelos disponibles en Ollama
def obtener_modelos_ollama():
    """Obtiene la lista de modelos disponibles en Ollama."""
    import subprocess
    try:
        # Ejecutar comando 'ollama list' para obtener modelos disponibles
        resultado = subprocess.run(['ollama', 'list'], capture_output=True, text=True)
        if resultado.returncode != 0:
            print(f"‚ùå Error al ejecutar 'ollama list': {resultado.stderr}")
            return []
        
        # Procesar la salida para obtener nombres de modelos
        lineas = resultado.stdout.strip().split('\n')
        if len(lineas) <= 1:  # Solo la cabecera, no hay modelos
            return []
            
        modelos = []
        for linea in lineas[1:]:  # Saltar la cabecera
            partes = linea.split()
            if partes:
                modelos.append(partes[0])  # El nombre es la primera columna
                
        return modelos
    except Exception as e:
        print(f"‚ùå Error al obtener modelos de Ollama: {str(e)}")
        return []

# Funci√≥n para verificar modelos de embeddings disponibles
def verificar_modelos_embeddings():
    """Verifica qu√© modelos de embeddings est√°n disponibles y retorna el primero que funcione."""
    # Si existe el archivo con el modelo guardado, lo usamos
    if os.path.exists(MODELO_EMBEDDINGS_FILE):
        try:
            with open(MODELO_EMBEDDINGS_FILE, 'r') as f:
                modelo_guardado = f.read().strip()
                print(f"Se encontr√≥ un modelo de embeddings guardado: {modelo_guardado}")
                return modelo_guardado
        except Exception as e:
            print(f"‚ùå Error al leer el archivo de modelo embeddings: {str(e)}")
    
    # Si no, buscamos un modelo disponible
    print("Verificando modelos de embeddings disponibles en Ollama...")
    modelos_disponibles = obtener_modelos_ollama()
    
    if not modelos_disponibles:
        print("‚ùå No se pudieron obtener los modelos disponibles de Ollama")
        return None
        
    print(f"Modelos disponibles en Ollama: {', '.join(modelos_disponibles)}")
    
    # Probar los modelos de embeddings en orden
    for modelo in MODELOS_EMBEDDINGS:
        if modelo in modelos_disponibles:
            print(f"Probando modelo de embeddings: {modelo}")
            try:
                # Intenta crear embeddings con este modelo
                embeddings = OllamaEmbeddings(
                    model=modelo,
                    base_url="http://localhost:11434",
                )
                # Probar el modelo con un texto simple
                test_embed = embeddings.embed_query("Prueba de embeddings")
                if len(test_embed) > 0:
                    print(f"‚úÖ Modelo de embeddings {modelo} funciona correctamente")
                    return modelo
            except Exception as e:
                print(f"‚ùå Error al probar modelo {modelo}: {str(e)}")
    
    # Si llegamos aqu√≠, ning√∫n modelo funcion√≥
    print("‚ùå Ninguno de los modelos de embeddings conocidos est√° disponible")
    print("Modelos que intentamos usar:")
    for modelo in MODELOS_EMBEDDINGS:
        print(f"  - {modelo}")
    return None

# Funci√≥n para cargar el √≠ndice vectorial y metadatos
def cargar_indice_vectorial():
    """Carga el √≠ndice vectorial FAISS y los metadatos de documentos."""
    global documentos_completos
    
    print("=== CARGANDO √çNDICE VECTORIAL ===")
    
    # Verificar que existan los archivos necesarios
    if not os.path.exists(FAISS_INDEX_PATH):
        print(f"‚ùå ERROR: No se encontr√≥ el √≠ndice FAISS en {FAISS_INDEX_PATH}")
        print("Ejecuta primero el script de creaci√≥n de √≠ndice.")
        return None, None
    
    if not os.path.exists(METADATA_FILE):
        print(f"‚ùå ERROR: No se encontr√≥ el archivo de metadatos en {METADATA_FILE}")
        return None, None
    
    try:
        # Cargar metadatos de documentos
        with open(METADATA_FILE, 'r', encoding='utf-8') as f:
            metadata = json.load(f)
        print(f"‚úÖ Metadatos de documentos cargados: {len(metadata)} documentos")
        
        # Cargar los documentos completos si existen
        if os.path.exists(DOCUMENTOS_FILE):
            with open(DOCUMENTOS_FILE, 'rb') as f:
                # Permitir deserializaci√≥n porque es un archivo que t√∫ mismo creaste
                docs_list = pickle.load(f)
            print(f"‚úÖ Lista de documentos cargada: {len(docs_list)} fragmentos")
            
            # Organizar documentos por ruta para acceso r√°pido
            documentos_por_ruta = {}
            for doc in docs_list:
                ruta = doc.metadata.get('source')
                if ruta not in documentos_por_ruta:
                    documentos_por_ruta[ruta] = []
                documentos_por_ruta[ruta].append(doc)
            
            documentos_completos = documentos_por_ruta
        else:
            print("‚ö†Ô∏è No se encontr√≥ el archivo de documentos completos, no se podr√° recuperar documentos enteros")
        
        # Obtener modelo de embeddings
        modelo_embeddings = verificar_modelos_embeddings()
        if not modelo_embeddings:
            print("‚ùå No se encontr√≥ un modelo de embeddings v√°lido.")
            return None, None
            
        # Cargar embeddings
        print(f"Configurando embeddings con modelo: {modelo_embeddings}...")
        embeddings = OllamaEmbeddings(
            model=modelo_embeddings,
            base_url="http://localhost:11434",
        )
        
        # Cargar el √≠ndice FAISS con el par√°metro allow_dangerous_deserialization
        print("Cargando √≠ndice FAISS...")
        try:
            vectorstore = FAISS.load_local(
                FAISS_INDEX_PATH, 
                embeddings,
                allow_dangerous_deserialization=True  # ¬°IMPORTANTE! A√±adir este par√°metro
            )
            print(f"‚úÖ √çndice FAISS cargado correctamente")
            return vectorstore, metadata
        except Exception as e:
            print(f"‚ùå Error al cargar el √≠ndice FAISS: {str(e)}")
            print(f"Detalles: {traceback.format_exc()}")
            return None, None
        
    except Exception as e:
        print(f"‚ùå Error al cargar el √≠ndice vectorial: {str(e)}")
        print(f"Detalles: {traceback.format_exc()}")
        return None, None

# Verificar Ollama y cargar el √≠ndice
def inicializar_sistema():
    """Verifica que Ollama est√© disponible y carga el √≠ndice vectorial."""
    # Verificar que Ollama est√© en ejecuci√≥n
    modelos_disponibles = obtener_modelos_ollama()
    if not modelos_disponibles:
        print("‚ùå No se pudo conectar con Ollama o no hay modelos disponibles.")
        return None, None
        
    if MODELO_OLLAMA not in modelos_disponibles:
        print(f"‚ùå El modelo LLM '{MODELO_OLLAMA}' no est√° disponible en Ollama.")
        print(f"Modelos disponibles: {', '.join(modelos_disponibles)}")
        print(f"Por favor, instala el modelo con: ollama pull {MODELO_OLLAMA}")
        return None, None
    
    try:
        test_model = Ollama(model=MODELO_OLLAMA)
        test_response = test_model.invoke("Prueba de conexi√≥n")
        print(f"‚úÖ Conexi√≥n con Ollama establecida correctamente usando modelo {MODELO_OLLAMA}")
    except Exception as e:
        print(f"‚ùå Error al conectar con Ollama: {str(e)}")
        print("Aseg√∫rate de que Ollama est√© en ejecuci√≥n y que los modelos necesarios est√©n instalados.")
        print(f"Puedes instalar el modelo LLM necesario con:")
        print(f"  ollama pull {MODELO_OLLAMA}")
        return None, None
    
    # Cargar el √≠ndice vectorial
    return cargar_indice_vectorial()

# Funci√≥n segura para invocar Ollama
def invocar_ollama_seguro(prompt, model=MODELO_OLLAMA, temperature=0):
    """Invoca Ollama con manejo de errores robusto."""
    try:
        modelo = Ollama(model=model, temperature=temperature)
        response = modelo.invoke(prompt)
        return response
    except Exception as e:
        print(f"‚ö†Ô∏è Error al invocar Ollama: {str(e)}")
        # Intenta un prompt m√°s corto si es muy largo
        if len(prompt) > 8192:
            print("Intentando con un prompt m√°s corto...")
            return invocar_ollama_seguro(prompt[:8000] + "...", model, temperature)
        return f"Error al generar respuesta: {str(e)}"

# Inicializaci√≥n principal
try:
    print("Inicializando sistema...")
    vectorstore, metadata_docs = inicializar_sistema()

    if vectorstore is None or metadata_docs is None:
        print("No se pudo cargar el √≠ndice vectorial. El programa se cerrar√°.")
        import sys
        sys.exit(1)

    # Crear el recuperador
    retriever = vectorstore.as_retriever(
        search_type="similarity",
        search_kwargs={"k": 5}  # Recuperar los 5 documentos m√°s relevantes
    )
except Exception as e:
    print(f"‚ùå Error durante la inicializaci√≥n: {str(e)}")
    print(f"Detalles: {traceback.format_exc()}")
    import sys
    sys.exit(1)

# Definir herramienta de b√∫squeda
@tool
def buscar_documentos(query: str) -> str:
    """Busca informaci√≥n en documentos PDF basada en la consulta proporcionada."""
    try:
        docs = retriever.get_relevant_documents(query)
        if not docs:
            return "No se encontraron documentos relevantes para esta consulta."
        
        results = []
        for i, doc in enumerate(docs):
            source = doc.metadata.get("source", "Desconocido")
            page = doc.metadata.get("page", "Desconocida")
            nombre_archivo = os.path.basename(source) if source != "Desconocido" else "Desconocido"
            results.append(f"Documento {i+1} (Archivo: {nombre_archivo}, P√°gina: {page}):\n{doc.page_content}\n")
        
        return "\n".join(results)
    except Exception as e:
        return f"Error durante la b√∫squeda: {str(e)}"

# Definir herramienta para recuperar documentos completos
@tool
def recuperar_documento_completo(nombre_documento: str) -> str:
    """Recupera el documento completo con el nombre especificado."""
    try:
        # Buscar el documento por nombre en los metadatos
        documentos_coincidentes = []
        
        for ruta in metadata_docs.keys():
            nombre_archivo = os.path.basename(ruta)
            if nombre_documento.lower() in nombre_archivo.lower():
                documentos_coincidentes.append(ruta)
        
        if not documentos_coincidentes:
            return f"No se encontr√≥ ning√∫n documento con el nombre '{nombre_documento}'. Documentos disponibles: {', '.join([os.path.basename(ruta) for ruta in metadata_docs.keys()])}"
        
        if len(documentos_coincidentes) > 1:
            nombres = [os.path.basename(ruta) for ruta in documentos_coincidentes]
            return f"Encontrados m√∫ltiples documentos coincidentes: {', '.join(nombres)}. Por favor, especifica el nombre con m√°s detalle."
        
        # Si solo hay una coincidencia, recuperar el documento completo
        ruta = documentos_coincidentes[0]
        
        # Verificar si tenemos el documento completo cargado
        if ruta not in documentos_completos:
            return f"El documento {os.path.basename(ruta)} est√° en los metadatos pero no fue cargado completamente. Por favor, ejecuta nuevamente el script de creaci√≥n de √≠ndice."
        
        paginas = documentos_completos[ruta]
        paginas_ordenadas = sorted(paginas, key=lambda x: int(x.metadata.get('page', 0)))
        
        # Formatear el contenido del documento completo
        contenido = f"DOCUMENTO COMPLETO: {os.path.basename(ruta)}\n"
        contenido += f"Ruta: {ruta}\n"
        contenido += f"N√∫mero de p√°ginas: {len(paginas_ordenadas)}\n\n"
        
        for i, pagina in enumerate(paginas_ordenadas):
            contenido += f"--- P√ÅGINA {i+1} ---\n"
            contenido += pagina.page_content
            contenido += "\n\n"
        
        return contenido
    except Exception as e:
        return f"Error al recuperar documento: {str(e)}"

# Definir herramienta para listar documentos disponibles
@tool
def listar_documentos_disponibles() -> str:
    """Lista todos los documentos PDF disponibles en el sistema."""
    try:
        if not metadata_docs:
            return "No hay documentos disponibles en el sistema."
        
        resultado = "Documentos disponibles:\n\n"
        for ruta, info in metadata_docs.items():
            nombre = os.path.basename(ruta)
            paginas = info.get('paginas', 'Desconocido')
            resultado += f"- {nombre} ({paginas} p√°ginas)\n"
        
        return resultado
    except Exception as e:
        return f"Error al listar documentos: {str(e)}"

# Lista de herramientas disponibles
tools = [buscar_documentos, recuperar_documento_completo, listar_documentos_disponibles]

# Definici√≥n del estado del agente
class AgentState(TypedDict):
    messages: Annotated[Sequence[BaseMessage], add_messages]

# Funci√≥n para determinar si una consulta es sobre un documento completo
def es_consulta_documento_completo(query):
    """Determina si la consulta solicita un documento completo."""
    keywords = ["documento completo", "archivo entero", "pdf completo", 
               "texto √≠ntegro", "todas las p√°ginas", "documento entero",
               "recupera el documento", "dame el archivo", "obt√©n el pdf"]
    
    return any(keyword in query.lower() for keyword in keywords)

# Funci√≥n para extraer el nombre del documento de la consulta
def extraer_nombre_documento(query):
    """Extrae el nombre del documento de la consulta."""
    palabras = query.split()
    # Buscar palabras despu√©s de "documento", "archivo", "pdf"
    for i, palabra in enumerate(palabras):
        if palabra.lower() in ["documento", "archivo", "pdf"] and i < len(palabras) - 1:
            # Tomar el resto de la consulta como posible nombre
            return " ".join(palabras[i+1:])
    
    # Si no encontramos palabras clave espec√≠ficas, eliminar palabras comunes
    keywords = ["documento", "completo", "archivo", "entero", "pdf", "dame", "recupera", "obt√©n", "el", "la", "los", "las"]
    return " ".join([p for p in palabras if p.lower() not in keywords]).strip()

# Funci√≥n mejorada para determinar si una consulta necesita b√∫squeda
def necesita_busqueda(query):
    """Determina si la consulta necesita buscar informaci√≥n en los documentos."""
    # Si solicita listar documentos
    if "lista" in query.lower() and "documentos" in query.lower():
        return False
        
    # Si solicita un documento completo
    if es_consulta_documento_completo(query):
        return False
    
    # Lista ampliada de palabras clave interrogativas y de b√∫squeda
    keywords = ["busca", "encuentra", "informaci√≥n", "documento", "pdf", 
               "qu√©", "cu√°l", "c√≥mo", "por qu√©", "cu√°ndo", "d√≥nde", 
               "explica", "describe", "analiza", "compara", "enumera", "lista", "detalla",
               "menciona", "indica", "muestra", "se√±ala", "espec√≠fica"]
    
    # Verificar si la consulta contiene alguna palabra clave
    contiene_keyword = any(keyword in query.lower() for keyword in keywords)
    
    # Verificar si parece una pregunta
    es_pregunta = '?' in query or any(query.lower().startswith(w) for w in ["qu√©", "cu√°l", "c√≥mo", "por", "qui√©n", "cu√°ndo", "d√≥nde"])
    
    return contiene_keyword or es_pregunta

# Funci√≥n del agente principal
def agent(state):
    """Procesa la entrada del usuario y decide qu√© acci√≥n tomar."""
    try:
        print("---LLAMANDO AL AGENTE---")
        messages = state["messages"]
        query = messages[0].content
        
        # Comando especial para listar documentos
        if "lista" in query.lower() and "documentos" in query.lower():
            print("El agente lista los documentos disponibles")
            return {
                "messages": [
                    ToolMessage(
                        content=listar_documentos_disponibles(),
                        name="listar_documentos_disponibles",
                        tool_call_id=str(uuid.uuid4())
                    )
                ]
            }
        
        # Verificar si es una solicitud de documento completo
        if es_consulta_documento_completo(query):
            print("El agente decide recuperar un documento completo")
            nombre_documento = extraer_nombre_documento(query)
            return {
                "messages": [
                    ToolMessage(
                        content=recuperar_documento_completo(nombre_documento),
                        name="recuperar_documento_completo",
                        tool_call_id=str(uuid.uuid4())
                    )
                ]
            }
        
        # Verificar si necesita b√∫squeda normal
        if necesita_busqueda(query):
            print("El agente decide usar la herramienta de b√∫squeda")
            return {
                "messages": [
                    ToolMessage(
                        content=buscar_documentos(query),
                        name="buscar_documentos",
                        tool_call_id=str(uuid.uuid4())
                    )
                ]
            }
        else:
            # Para consultas m√°s simples, responder directamente
            print("El agente responde directamente")
            response = invocar_ollama_seguro(f"Responde a la siguiente pregunta de manera concisa: {query}")
            return {"messages": [AIMessage(content=response)]}
    except Exception as e:
        print(f"Error en el agente: {str(e)}")
        print(traceback.format_exc())
        return {"messages": [AIMessage(content=f"Lo siento, ocurri√≥ un error al procesar tu consulta: {str(e)}")]}

# Funci√≥n para evaluar la relevancia de los documentos
def grade_documents(state):
    """Eval√∫a si los documentos recuperados son relevantes para la consulta."""
    try:
        print("---EVALUANDO RELEVANCIA DE DOCUMENTOS---")
        messages = state["messages"]
        last_message = messages[-1]
        question = messages[0].content
        
        # Si el √∫ltimo mensaje es un ToolMessage, extraemos su contenido
        docs = last_message.content
        
        # Si es un documento completo o lista de documentos, lo consideramos relevante
        if "DOCUMENTO COMPLETO:" in docs or "Documentos disponibles:" in docs:
            print("---DECISI√ìN: DOCUMENTO COMPLETO O LISTA DE DOCUMENTOS---")
            return "generate"
        
        # Prompt para evaluar relevancia
        prompt = f"""Est√°s evaluando la relevancia de un documento recuperado para una pregunta del usuario.
        
        Aqu√≠ est√° el documento recuperado:
        {docs[:1000]}... (documento truncado)
        
        Aqu√≠ est√° la pregunta del usuario:
        {question}
        
        Si el documento contiene palabras clave o significado sem√°ntico relacionado con la pregunta del usuario, calif√≠calo como relevante.
        Da una puntuaci√≥n binaria 'si' o 'no' para indicar si el documento es relevante para la pregunta.
        
        Responde solo con 'si' o 'no'.
        """
        
        # Verificar relevancia
        result = invocar_ollama_seguro(prompt, MODELO_OLLAMA, 0)
        score = "si" if "si" in result.lower() else "no"
        
        if score == "si":
            print("---DECISI√ìN: DOCUMENTOS RELEVANTES---")
            return "generate"
        else:
            print("---DECISI√ìN: DOCUMENTOS NO RELEVANTES---")
            print(f"Resultado de la evaluaci√≥n: {result}")
            return "rewrite"
    except Exception as e:
        print(f"Error en evaluaci√≥n de documentos: {str(e)}")
        print(traceback.format_exc())
        return "generate"  # En caso de error, intentamos generar una respuesta

# Funci√≥n para reformular la consulta
def rewrite(state):
    """Transforma la consulta para producir una mejor pregunta."""
    try:
        print("---TRANSFORMANDO CONSULTA---")
        messages = state["messages"]
        question = messages[0].content

        # Prompt para mejorar la pregunta
        prompt_text = f"""
        Mira la siguiente consulta e intenta entender su intenci√≥n sem√°ntica/significado subyacente.
        Aqu√≠ est√° la pregunta inicial:
        -------
        {question} 
        -------
        Formula una pregunta mejorada que ayude a recuperar informaci√≥n m√°s relevante de documentos PDF.
        La nueva pregunta debe ser clara, espec√≠fica y orientada a encontrar informaci√≥n en documentos.
        No expliques la reformulaci√≥n, simplemente proporciona la nueva pregunta.
        """
        
        # Usar Ollama para reformular
        response_text = invocar_ollama_seguro(prompt_text, MODELO_OLLAMA, 0)
        
        # Crear mensaje de respuesta
        response = HumanMessage(content=response_text)
        return {"messages": [response]}
    except Exception as e:
        print(f"Error al reformular consulta: {str(e)}")
        print(traceback.format_exc())
        # Si hay un error, intentamos con la consulta original
        return {"messages": [HumanMessage(content=question)]}

# Funci√≥n para generar la respuesta final
def generate(state):
    """Genera la respuesta final basada en los documentos recuperados."""
    try:
        print("---GENERANDO RESPUESTA---")
        messages = state["messages"]
        question = messages[0].content
        last_message = messages[-1]
        docs = last_message.content

        # Si es un documento completo o lista de documentos, lo devolvemos directamente
        if "DOCUMENTO COMPLETO:" in docs or "Documentos disponibles:" in docs:
            return {"messages": [AIMessage(content=f"Aqu√≠ est√° la informaci√≥n solicitada:\n\n{docs}")]}

        # Prompt para RAG mejorado
        rag_prompt = f"""
        Eres un asistente para tareas de respuesta a preguntas. Utiliza la siguiente informaci√≥n recuperada para responder a la pregunta.
        Si no conoces la respuesta basada en la informaci√≥n proporcionada, simplemente di que no puedes encontrar informaci√≥n suficiente.
        
        Pregunta: {question} 
        
        Informaci√≥n de los documentos:
        {docs} 
        
        Proporciona una respuesta clara, concisa y bien estructurada. Utiliza solo la informaci√≥n de los documentos proporcionados.
        Si citas informaci√≥n espec√≠fica, menciona de qu√© documento proviene.
        """
        
        # Generar respuesta
        response = invocar_ollama_seguro(rag_prompt, MODELO_OLLAMA, 0.1)
        
        return {"messages": [AIMessage(content=response)]}
    except Exception as e:
        print(f"Error al generar respuesta: {str(e)}")
        print(traceback.format_exc())
        return {"messages": [AIMessage(content=f"Lo siento, ocurri√≥ un error al generar la respuesta: {str(e)}")]}

# Simplificar la l√≥gica condicional
def route_agent_output(state):
    """Determina la siguiente acci√≥n basada en el tipo de mensaje."""
    try:
        messages = state["messages"]
        last_message = messages[-1]
        
        # Si el √∫ltimo mensaje no es un ToolMessage, significa que el agente respondi√≥ directamente
        if not isinstance(last_message, ToolMessage):
            return END
        
        # Si es un documento completo o lista de documentos, vamos directamente a generar
        if "DOCUMENTO COMPLETO:" in last_message.content or "Documentos disponibles:" in last_message.content:
            return "generate"
        
        # Evaluar relevancia de documentos para b√∫squedas normales
        return grade_documents(state)
    except Exception as e:
        print(f"Error en la funci√≥n de enrutamiento: {str(e)}")
        print(traceback.format_exc())
        return "generate"  # En caso de error, intentamos generar una respuesta

# Configurar el flujo de trabajo
try:
    print("Configurando el flujo de trabajo...")
    workflow = StateGraph(AgentState)

    # A√±adir nodos
    workflow.add_node("agent", agent)
    workflow.add_node("rewrite", rewrite) 
    workflow.add_node("generate", generate)

    # Configurar aristas
    workflow.add_edge(START, "agent")

    # A√±adir arista condicional para la salida del agente
    workflow.add_conditional_edges(
        "agent",
        route_agent_output,
        {
            END: END,
            "generate": "generate",
            "rewrite": "rewrite"
        }
    )

    workflow.add_edge("generate", END)
    workflow.add_edge("rewrite", "agent")

    # Compilar el grafo
    print("Compilando el grafo...")
    graph = workflow.compile()
except Exception as e:
    print(f"‚ùå Error al configurar el flujo de trabajo: {str(e)}")
    print(traceback.format_exc())
    import sys
    sys.exit(1)

# Funci√≥n para ejecutar una consulta
def ejecutar_consulta(pregunta):
    try:
        print(f"\n\n=== NUEVA CONSULTA: {pregunta} ===\n")
        inputs = {
            "messages": [
                HumanMessage(content=pregunta),
            ]
        }
        
        # Ejecutar el flujo completo
        result = graph.invoke(inputs)
        
        # Obtener la respuesta final
        respuesta_final = result["messages"][-1].content if isinstance(result["messages"][-1], BaseMessage) else result["messages"][-1]
        
        print("\n=== RESPUESTA FINAL ===")
        print(respuesta_final)
        return respuesta_final
    except Exception as e:
        print(f"Error al procesar la consulta: {str(e)}")
        print(traceback.format_exc())
        return f"Error: {str(e)}"

# Funci√≥n principal
if __name__ == "__main__":
    print("\n" + "="*50)
    print("AGENTE RAG PARA CONSULTA DE DOCUMENTOS PDF")
    print("="*50 + "\n")
    
    print("Sistema inicializado y listo para consultas.")
    print("\nComandos √∫tiles:")
    print("- \"lista documentos disponibles\": Muestra todos los documentos cargados")
    print("- \"dame el documento completo [nombre]\": Recupera un documento completo")
    print("- \"salir\": Termina el programa")
    
    # Ejecutar una consulta de ejemplo
    pregunta_ejemplo = "¬øCu√°les son los principales temas tratados en los documentos?"
    print(f"\nEjecutando consulta de ejemplo: \"{pregunta_ejemplo}\"")
    ejecutar_consulta(pregunta_ejemplo)
    
    # Modo interactivo
    print("\n\n=== MODO INTERACTIVO ===")
    print("Puedes hacer preguntas sobre tus documentos PDF. Escribe 'salir' para terminar.")
    
    while True:
        try:
            pregunta = input("\nTu pregunta: ")
            if pregunta.lower() in ["salir", "exit", "quit", "q"]:
                break
            
            ejecutar_consulta(pregunta)
        except KeyboardInterrupt:
            print("\nSaliendo del programa...")
            break
        except Exception as e:
            print(f"Error inesperado: {str(e)}")
            print(traceback.format_exc())
````

````python
import os
from datetime import datetime
import glob
import re

# 1. Guardar el documento con el formato espec√≠fico
def guardar_documento(contenido):
    # Crear carpeta si no existe
    if not os.path.exists("propuestas"):
        os.makedirs("propuestas")
    
    # Obtener fecha y hora actual en formato espec√≠fico
    current_date = datetime.now().strftime("%Y%m%d%H%M%S")
    
    # Crear nombre de archivo con el formato espec√≠fico
    nombre_archivo = f"Propuesta_TDR_{current_date}.txt"
    ruta_completa = os.path.join("propuestas", nombre_archivo)
    
    # Guardar el archivo
    with open(ruta_completa, "w") as f:
        f.write(contenido)
    
    return ruta_completa

# 2. Buscar archivo con la hora m√°s cercana a la hora actual
def buscar_archivo_mas_cercano():
    # Obtener hora actual
    hora_actual = datetime.now()
    
    # Buscar todos los archivos con el formato espec√≠fico
    patron_busqueda = os.path.join("propuestas", "Propuesta_TDR_*.txt")
    archivos = glob.glob(patron_busqueda)
    
    mejor_archivo = None
    menor_diferencia = float('inf')
    
    # Patr√≥n para extraer la fecha y hora del nombre del archivo
    patron_fecha = re.compile(r'Propuesta_TDR_(\d{14})\.txt$')
    
    for archivo in archivos:
        # Extraer la fecha del nombre del archivo
        match = patron_fecha.search(archivo)
        if match:
            fecha_str = match.group(1)
            try:
                # Convertir a objeto datetime
                fecha_archivo = datetime.strptime(fecha_str, "%Y%m%d%H%M%S")
                
                # Calcular diferencia en segundos con la hora actual
                diferencia = abs((fecha_archivo - hora_actual).total_seconds())
                
                # Actualizar si encontramos una diferencia menor
                if diferencia < menor_diferencia:
                    menor_diferencia = diferencia
                    mejor_archivo = archivo
            except ValueError:
                continue
    
    return mejor_archivo

# Ejemplo de uso
if __name__ == "__main__":
    # Guardar un documento de ejemplo
    contenido = "Este es un documento de prueba.2323"
    ruta_guardado = guardar_documento(contenido)
    print(f"Documento guardado en: {ruta_guardado}")
    
    # Buscar documento m√°s cercano a la hora actual
    archivo_cercano = buscar_archivo_mas_cercano()
    if archivo_cercano:
        print(f"Archivo m√°s cercano a la hora actual: {archivo_cercano}")
        # Leer el contenido
        try:
            with open(archivo_cercano, "r", encoding="utf-8") as f:
                contenido = f.read()
                print(f"Contenido: {contenido}")
        except UnicodeDecodeError:
            # Si falla con UTF-8, intentar en modo binario
            with open(archivo_cercano, "rb") as f:
                contenido = f.read()
                print(f"Contenido (binario): {contenido}")
    else:
        print("No se encontr√≥ ning√∫n archivo.")
````
